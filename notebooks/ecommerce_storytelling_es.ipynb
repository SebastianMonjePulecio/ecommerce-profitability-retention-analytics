```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd()
TABLES_DIR = BASE_DIR / 'outputs' / 'tables'

summary = pd.read_csv(TABLES_DIR / 'executive_summary.csv')
channels = pd.read_csv(TABLES_DIR / 'channel_performance.csv')
categories = pd.read_csv(TABLES_DIR / 'category_performance.csv')
rfm = pd.read_csv(TABLES_DIR / 'rfm_segments.csv')

summary
```

## 2. KPIs principales

Antes de entrar en segmentos o categor?as, conviene mirar la foto general del negocio. Estos indicadores ayudan a responder si el problema es de volumen, de margen o de eficiencia comercial.


```python
summary_map = dict(zip(summary['metric'], summary['value']))
kpis = pd.DataFrame({
    'KPI': ['Ingreso neto', 'Utilidad bruta', 'Margen bruto %', 'Ticket promedio', 'Tasa de recompra %', 'Tasa de devoluciones %', 'ROAS'],
    'Valor': [
        round(summary_map['net_revenue'], 2),
        round(summary_map['gross_profit'], 2),
        round(summary_map['gross_margin_pct'], 2),
        round(summary_map['average_order_value'], 2),
        round(summary_map['repeat_purchase_rate_pct'], 2),
        round(summary_map['return_rate_pct'], 2),
        round(summary_map['roas'], 2),
    ]
})
kpis
```

Cuando se analiza el desempe?o por canal, aparece una tensi?n muy normal en e-commerce: algunos canales traen ventas, pero no necesariamente dejan la mejor utilidad.


```python
channels[['acquisition_channel', 'net_revenue', 'gross_profit', 'gross_margin_pct', 'return_rate_pct', 'roas']]
    .sort_values('gross_profit', ascending=False)
```

Lectura de negocio:

- `Organic` y `Email` muestran una combinaci?n m?s saludable entre utilidad, margen y ROAS.
